```python
from sklearn.preprocessing import OneHotEncoder

# Assuming 'race' is the column you want to one-hot encode in your DataFrame 'df'
column_to_encode = 'race'

# Extract the column to be encoded
column_data = df_1[[column_to_encode]]

# Create the OneHotEncoder instance
encoder = OneHotEncoder()

# Fit and transform the data
encoded_features = encoder.fit_transform(column_data)

# Convert the sparse matrix to a DataFrame with column names
encoded_df = pd.DataFrame(encoded_features.toarray(), columns=encoder.get_feature_names_out([column_to_encode]))

# Concatenate the original DataFrame 'df' and the one-hot encoded DataFrame 'encoded_df'
df = pd.concat([df_1, encoded_df], axis=1)

# Drop the original column that was one-hot encoded
df.drop(column_to_encode, axis=1, inplace=True)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Assuming 'df' is your DataFrame
# Replace it with your actual DataFrame variable

# List of diabetes medications
diabetes_meds = ['metformin', 'repaglinide', 'nateglinide', 'chlorpropamide', 'glimepiride', 'acetohexamide',
                 'glipizide', 'glyburide', 'tolbutamide', 'pioglitazone', 'rosiglitazone', 'acarbose',
                 'miglitol', 'troglitazone', 'tolazamide', 'examide', 'citoglipton', 'glyburide-metformin',
                 'glipizide-metformin', 'glimepiride-pioglitazone', 'metformin-rosiglitazone', 'metformin-pioglitazone']

# Filter the DataFrame for relevant columns
df_diabetes_meds = df[diabetes_meds + ['insulin']]

# Set the style for better visualization
sns.set(style="whitegrid")

# Loop through each medication and create a grouped bar plot
for med in diabetes_meds:
    # Calculate the counts of combinations
    med_insulin_counts = df_diabetes_meds.groupby([med, 'insulin']).size().unstack(fill_value=0)

    # Plot a grouped bar chart
    med_insulin_counts.plot(kind='bar', figsize=(12, 6))
    
    # Set plot labels and title
    plt.xlabel(f'{med} and Insulin')
    plt.ylabel('Count')
    plt.title(f'Relationship between {med} and Insulin')

    # Show the plot
    plt.show()

```
