```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Assuming 'df' is your main dataset
data = {'airline': ['SpiceJet', 'AirAsia', 'Vistara', 'GO FIRST', 'Indigo', 'Air India', 'Trujet', 'StarAir'],
        'to': ['Mumbai', 'Bangalore', 'Kolkata', 'Hyderabad', 'Chennai', 'Delhi', 'SomeCity', 'AnotherCity'],
        'from': ['Delhi', 'Mumbai', 'Bangalore', 'Kolkata', 'Hyderabad', 'Chennai' , 'from1' , 'from2']}

df = pd.DataFrame(data)

# Extract the columns you want to encode
columns_to_encode = ['airline', 'to', 'from']

# Create a OneHotEncoder instance
encoder = OneHotEncoder(sparse=False, drop='first')

# Fit and transform the data
encoded_features = encoder.fit_transform(df[columns_to_encode])

# Get the feature names for the one-hot encoded columns
feature_names = encoder.get_feature_names_out(columns_to_encode)

# Create a DataFrame with the one-hot encoded features and column names
encoded_df = pd.DataFrame(encoded_features, columns=feature_names)

# Concatenate the one-hot encoded features with the original DataFrame
df = pd.concat([df, encoded_df], axis=1)

# Drop the original categorical columns if needed
df = df.drop(columns=columns_to_encode)

print(df)
```
