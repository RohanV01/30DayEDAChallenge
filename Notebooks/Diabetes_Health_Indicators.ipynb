```python
import pandas as pd 
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
```

```python
df_1 = pd.read_csv("C:\\C NEW\\30Days_EDAChallenege\\datasets\\diabetes_012_health_indicators_BRFSS2015.csv")
```

```python
df_1
```

```python
sns.barplot(x='Diabetes_012', y='Age', data=df_1)
plt.show()
```

#### From the above graph , I conclude the following - 
1. Children with ages more than 8 are pre diabetic and diabetic
2. Children who are bloew the age of 8 are less prone to be diabetic

```python
sns.countplot(x='Sex', hue='Diabetes_012', data=df_1)
plt.show()
```

#### From the above graph,I conclude the following - 
1. Males are more prone to diabetes than females

```python
sns.countplot(x='Diabetes_012', hue='Fruits', data=df_1)
plt.show()
```

#### From the above graph,I conclude the following - 
1. People who eat fruits are not prone to diabetes
2. Consuming fruits has no effect on prediabetes
3. There is a subset of population who is suffering from diabetes even though they are consuming fruits

```python
g = sns.FacetGrid(df_1, col='Diabetes_012')
g.map(plt.hist, 'BMI', bins=20)
plt.show()
```

#### From the above graph,I conclude the following - 

1. People with BMI in the range of 20-40 are not prone to diabetes
2. People in the BMI range of 20-40 are also prone to diabetes -> This probably has something to do with Cholesterol and BP issues
